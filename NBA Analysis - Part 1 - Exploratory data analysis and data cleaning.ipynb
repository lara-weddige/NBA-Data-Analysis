# 3. Data wrangling

```python
# Changing data type of age from float to integer

df['age'] = df['age'].astype(int)
```

```python
df.dtypes
```

```python
# Looking at all the entries for country

unique_countries = df['country'].unique()
print(unique_countries)
```

```python
# Making the countries easier to read

unique_countries = sorted(df['country'].unique())
print('\n'.join(unique_countries))
```

```python
# Checking for missing values

df.isnull().sum()
```

The only missing values are in the college column. After cross-checking this information with certain associated player names, it appears this data is in fact complete and correct, as not all NBA players played in college prior to joining the NBA (could have skipped college entirely or been playing professionally outside of the US before being drafted). 

```python
# Find duplicates

df_dups = df[df.duplicated()]
```

```python
df_dups.shape
```

There are no duplicates in this dataset, so this does not need to be addressed further.

```python
# Checking for mixed datatypes

for col in df.columns.tolist():
      weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (df[weird]) > 0:
        print (col)
```

```python
# Changing all entries in college column to string

df['college'] = df['college'].astype(str)
```

```python
# Checking for mixed datatypes again

for col in df.columns.tolist():
      weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (df[weird]) > 0:
        print (col)
```

No more mixed data types! Great!

```python
# Replace missing values with 'Unknown' in college column

df['college'] = df['college'].fillna('Unknown')
```

```python
# Creating a new column indicating whether a player attended college

df['attended_college'] = df['college'].notnull()
```
